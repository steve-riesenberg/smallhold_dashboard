```python
import pandas as pd
import datetime
import plotly.express as px
```

```python
co2 = pd.read_csv('data/co2.csv', header=None, names=['timestamp', 'value'])
```

```python
humidity = pd.read_csv('data/humidity.csv', header=None, names=['timestamp', 'value'])
```

```python
temp = pd.read_csv('data/temp.csv', header=None, names=['timestamp', 'value'])
```

```python
co2.timestamp.apply(lambda x: x[-1]).value_counts()
```

```python
co2['date_time'] = co2.timestamp.apply(lambda x: datetime.datetime.strptime(x, "%Y-%m-%dT%H:%M:%S.%fZ"))
humidity['date_time'] = humidity.timestamp.apply(lambda x: datetime.datetime.strptime(x, "%Y-%m-%dT%H:%M:%S.%fZ"))
temp['date_time'] = temp.timestamp.apply(lambda x: datetime.datetime.strptime(x, "%Y-%m-%dT%H:%M:%S.%fZ"))
```

```python
for df in [co2, humidity, temp]:
    print(df.date_time.min(), df.date_time.max())
```
